```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/diabetes_health_indicators.csv")
print("Dataset Shape: ", df.shape)
df.head()
```

```python
df["Diabetes_012"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Diabetes_012", data=df)
plt.title("Class Distribution (0=Healthy, 1=Pre, 2=Diabetic)")
plt.show()
```

1. Class 1 (Pre-diabetes) is extremely low
2. Dataset is highly imbalanced
3. Model will struggle with class 1

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Diabetes_012", y="BMI", data=df)
plt.title("BMI vs Diabetes Status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Diabetes_012", y="Age", data=df)
plt.title("Age vs Diabetes Status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Diabetes_012", y="GenHlth", data=df)
plt.title("General Health vs Diabetes")
plt.show()
```
